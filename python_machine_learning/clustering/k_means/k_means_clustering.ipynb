```python
import numpy as np
import pandas as pd
from ucimlrepo import fetch_ucirepo
import matplotlib.pyplot as plt 
```

## Importing Data and Preprocessing

```python
 # fetch cancer dataset 
breast_cancer_wisconsin_diagnostic = fetch_ucirepo(id=17) 
  
# data (as pandas dataframes) 
X_cancer = breast_cancer_wisconsin_diagnostic.data.features 
y_cancer = breast_cancer_wisconsin_diagnostic.data.targets 

# metadata 
# print(breast_cancer_wisconsin_diagnostic.metadata)
# variable information
# print(breast_cancer_wisconsin_diagnostic.variables)

# fetch waveform dataset 
waveform_database_generator_version_1 = fetch_ucirepo(id=107) 
  
# data (as pandas dataframes) 
X_waveform = waveform_database_generator_version_1.data.features 
y_waveform = waveform_database_generator_version_1.data.targets 
  
# metadata 
# print(waveform_database_generator_version_1.metadata)
# variable information 
# print(waveform_database_generator_version_1.variables)
```

```python
from sklearn.preprocessing import StandardScaler

X_cancer = pd.DataFrame(StandardScaler().fit_transform(X=X_cancer), columns=X_cancer.columns)
X_waveform = pd.DataFrame(StandardScaler().fit_transform(X=X_waveform), columns=X_waveform.columns)

X_cancer_feature_ranges = pd.DataFrame(X_cancer.apply(lambda x: x.max() - x.min(), axis=0), index=X_cancer.columns)
X_cancer_feature_ranges.plot(kind="bar", xlabel='feature', ylabel='range', title="Breast Cancer Feature Range",
                             xticks=range(X_cancer_feature_ranges.size), legend=False)
X_waveform_feature_ranges = pd.DataFrame(X_waveform.apply(lambda x: x.max() - x.min(), axis=0), index=X_waveform.columns)
X_waveform_feature_ranges.plot(kind="bar", xlabel='feature', ylabel='range', title="Waveform Feature Range",
                               xticks=range(X_waveform_feature_ranges.size), legend=False)
plt.show()
```

## Part 1

```python
from sklearn import cluster

cancer_k_values = np.arange(1,7)
cancer_sse = []
for k in cancer_k_values:
    _, __, sse = cluster.k_means(X=X_cancer, n_clusters=k, n_init=10, max_iter=300)
    cancer_sse.append(sse)


waveform_k_values = np.arange(2,7)
waveform_sse = []
for k in waveform_k_values:
    _, __, sse = cluster.k_means(X=X_waveform, n_clusters=k, n_init=10, max_iter=300)
    waveform_sse.append(sse)

plt.plot([1,2,3,4,5,6], cancer_sse, 'r--o')
plt.title('Elbow Method for Optimal k (Breast Cancer)')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('SSE')
plt.grid()
plt.show()

plt.plot([2,3,4,5,6], waveform_sse, 'r--o')
plt.title('Elbow Method for Optimal k (Waveform)')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('SSE')
plt.xticks([2,3,4,5,6])
plt.grid()
plt.show()
```

```python
display(y_cancer.value_counts())
display(y_waveform.value_counts())
```

```python
# distance measure is euclidean distance

def distCentroid(x, centroids):
    return np.sqrt(np.sum((centroids[int(x['cluster_ID'])]-x[x.index!='cluster_ID'])**2))

cancer_centroids, cancer_cluster, cancer_sse = cluster.k_means(X=X_cancer, n_clusters=2, n_init=10, max_iter=300)
X_cancer['cluster_ID'] = cancer_cluster
X_cancer['dist_centroid'] = X_cancer.apply(lambda x: distCentroid(x,cancer_centroids), axis=1)


waveform_centroids, waveform_cluster, waveform_sse = cluster.k_means(X=X_waveform, n_clusters=3, n_init=10, max_iter=300)
X_waveform['cluster_ID'] = waveform_cluster
X_waveform['dist_centroid'] = X_waveform.apply(lambda x: distCentroid(x, waveform_centroids), axis=1)
```

```python
print("Cancer Data Cluster Centroids (all dimensions):")
cancer_centroid_df = pd.DataFrame(cancer_centroids, columns=X_cancer.columns[:-2])
cancer_centroid_df.index.name = 'cluster_ID'
display(cancer_centroid_df.iloc[:,:10])
display(cancer_centroid_df.iloc[:,10:20])
display(cancer_centroid_df.iloc[:,20:])

print(f"Top five closest points to centeroids of\nBreast Cancer Dataset:")
top_points_cancer = pd.concat(
    [X_cancer.loc[X_cancer['cluster_ID']==k,:].sort_values(by='dist_centroid', ascending=True).iloc[:5,-2:] for k in range(2)],
    axis=0)
top_points_cancer.index.name = 'record_ID' 
display(top_points_cancer)

print("Waveform Data Cluster Centroids (all dimensions):")
waveform_centroid_df = pd.DataFrame(waveform_centroids, columns=X_waveform.columns[:-2])
waveform_centroid_df.index.name = 'cluster_ID'
display(waveform_centroid_df.iloc[:,:10])
display(waveform_centroid_df.iloc[:,10:])

print(f"Top five closest points to centeroids of\nWaveform Dataset:")
top_points_waveform = pd.concat(
    [X_waveform.loc[X_waveform['cluster_ID']==k,:].sort_values(by='dist_centroid', ascending=True).iloc[:5,-2:] for k in range(3)],
    axis=0)
top_points_waveform.index.name = 'record_ID' 
display(top_points_waveform)
```

## Part 2:

```python
from scipy.cluster import hierarchy
import numpy as np

# Linkage and dendrogram
Z = hierarchy.linkage(X_cancer.iloc[:,:-2].to_numpy(), 'single')
plt.title('Single Link Hierarchical Clustering Dendrogram (truncated)\nBreast Cancer Dataset')
plt.xlabel('sample index or (cluster size)')
plt.ylabel('distance')
hierarchy.dendrogram(
    Z,
    truncate_mode='lastp',  # show only the last p merged clusters
    p=20,  # show only the last p merged clusters
    leaf_rotation=90.,
    leaf_font_size=12.,
    show_contracted=True,  # to get a distribution impression in truncated branches
)
plt.show()
```

```python
# Linkage and dendrogram
X_waveform_sample = X_waveform.sample(n=4300)
Z = hierarchy.linkage(X_waveform_sample.iloc[:,:-2].to_numpy(), 'single')
plt.title('Single Link Hierarchical Clustering Dendrogram (truncated)\nWaveform Dataset')
plt.xlabel('sample index or (cluster size)')
plt.ylabel('distance')
hierarchy.dendrogram(
    Z,
    truncate_mode='lastp',  # show only the last p merged clusters
    p=30,  # show only the last p merged clusters
    leaf_rotation=90.,
    leaf_font_size=12.,
    show_contracted=True,  # to get a distribution impression in truncated branches
)
plt.show()
```
