```python
import numpy as np
import pandas as pd
from ucimlrepo import fetch_ucirepo
import matplotlib.pyplot as plt 
```

## Importing Data and Preprocessing

```python
 # fetch cancer dataset 
breast_cancer_wisconsin_diagnostic = fetch_ucirepo(id=17) 
  
# data (as pandas dataframes) 
X_cancer = breast_cancer_wisconsin_diagnostic.data.features 
y_cancer = breast_cancer_wisconsin_diagnostic.data.targets 

# metadata 
# print(breast_cancer_wisconsin_diagnostic.metadata)
# variable information
# print(breast_cancer_wisconsin_diagnostic.variables)

# fetch waveform dataset 
waveform_database_generator_version_1 = fetch_ucirepo(id=107) 
  
# data (as pandas dataframes) 
X_waveform = waveform_database_generator_version_1.data.features 
y_waveform = waveform_database_generator_version_1.data.targets 
  
# metadata 
# print(waveform_database_generator_version_1.metadata)
# variable information 
# print(waveform_database_generator_version_1.variables)
```

#### Feature Range Visualization

```python
X_cancer_feature_ranges = pd.DataFrame(X_cancer.apply(lambda x: x.max() - x.min(), axis=0), index=X_cancer.columns)
X_cancer_feature_ranges.plot(kind="bar", xlabel='feature', ylabel='range', title="Breast Cancer Feature Range",
                             xticks=range(X_cancer_feature_ranges.size), legend=False)
X_waveform_feature_ranges = pd.DataFrame(X_waveform.apply(lambda x: x.max() - x.min(), axis=0), index=X_waveform.columns)
X_waveform_feature_ranges.plot(kind="bar", xlabel='feature', ylabel='range', title="Waveform Feature Range",
                               xticks=range(X_waveform_feature_ranges.size), legend=False)

plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

X_cancer = pd.DataFrame(StandardScaler().fit_transform(X=X_cancer), columns=X_cancer.columns)
X_waveform = pd.DataFrame(StandardScaler().fit_transform(X=X_waveform), columns=X_waveform.columns)

X_cancer_feature_ranges = pd.DataFrame(X_cancer.apply(lambda x: x.max() - x.min(), axis=0), index=X_cancer.columns)
X_cancer_feature_ranges.plot(kind="bar", xlabel='feature', ylabel='range', title="Breast Cancer Feature Range",
                             xticks=range(X_cancer_feature_ranges.size), legend=False)
X_waveform_feature_ranges = pd.DataFrame(X_waveform.apply(lambda x: x.max() - x.min(), axis=0), index=X_waveform.columns)
X_waveform_feature_ranges.plot(kind="bar", xlabel='feature', ylabel='range', title="Waveform Feature Range",
                               xticks=range(X_waveform_feature_ranges.size), legend=False)
plt.show()
```

## Part 1

```python
from sklearn import cluster

cancer_k_values = np.arange(1,7)
cancer_sse = []
for k in cancer_k_values:
    _, __, sse = cluster.k_means(X=X_cancer, n_clusters=k, n_init=10, max_iter=300)
    cancer_sse.append(sse)


waveform_k_values = np.arange(2,7)
waveform_sse = []
for k in waveform_k_values:
    _, __, sse = cluster.k_means(X=X_waveform, n_clusters=k, n_init=10, max_iter=300)
    waveform_sse.append(sse)

plt.plot([1,2,3,4,5,6], cancer_sse, 'r--o')
plt.title('Elbow Method for Optimal k (Breast Cancer)')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('SSE')
plt.grid()
plt.show()

plt.plot([2,3,4,5,6], waveform_sse, 'r--o')
plt.title('Elbow Method for Optimal k (Waveform)')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('SSE')
plt.xticks([2,3,4,5,6])
plt.grid()
plt.show()
```
