```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure(figsize=(5,5))
ax = fig.gca()

# changing the labels
ax.set_xlabel("Products")
ax.set_ylabel("Sales")

# changing the xticks
ax.set_xticks(np.linspace(-10,10,9))
ax.set_xticklabels(
    [
        "cat",
        'dog',
        'bull',
        'cow',
        'sheep',
        'car',
        'monk',
        'hat',
        'lat'
    ]
)
plt.show()
```
