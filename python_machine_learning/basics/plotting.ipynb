```python
import matplotlib.pyplot as plt
import numpy as np
```

### 1 Matplotlib
---

- it is much more intuitive and sensible imo to use `plt.subplots((shape))` to first create a fig and axes object
    
    - the axes object can then be used for all plotting purposes

- too much reliance on pyplot's state-based interface (ex. using `plt.plot(x,y,...)` to plot immediately) complicates plot customization
    
    - some customizations such as `set_xticklabels()` require retrieval of the axes object using `plt.gca()` while others do not
    
    - it is ambiguous which customizations require axes objects

    - the best solution is to just create a fig and axs variable in the beginning and use the axs object moving forward

```python
# use plt.subplots to create an intuitive Object Oriented grid of axes objects
    # each axes object can be used for individual plotting
    # all axes objects fall within a parent container called the 'figure'

# create the initial figure and axes object    
fig1, axs1 = plt.subplots(1,1)
# set the size of figure object
fig1.set_size_inches(3,3)
# set the color of the entire figure (everything outside of the rectangular axes plotting region)
fig1.set_facecolor(color='#005488')


# setting the color of the axs object's plotting region (inside the inner rectangle)
axs1.set_facecolor(color='#005488')
# setting x/y axis labels and label colors
axs1.set_xlabel("Bro Hooly bro", color='#a7a7a7')
axs1.set_ylabel("Bro NAAAH bro", color='#a7a7a7')

# this retrieves a dictionary of the 'spines' of the axes object and sets the color of all axes spines ('top', 'bottom', 'left', 'right')
axs1.spines[:].set_color('#ffffff')

# setting the colors for the ticks along x and y axis
axs1.tick_params(axis='x', colors='#a7a7a7')
axs1.tick_params(axis='y', colors='#a7a7a7')

# setting the actual x ticks
    # if the list contains values >= 1, largest tick is placed at the right edge of the axis, remaining ticks are placed proportionally
    # if list contains values < 1, right edge of axis is assumed to be 1, all ticks are placed based on this scaling
# axs.set_xticks([0, 0.01, 0.3, 0.9]) # here the right edge of the axis has a value of 1, ticks are placed accordingly
axs1.set_xticks([0, 5, 30, 35, 15]) # here the right edge of the x axis has a value of max([0, 5, 30, 35, 15])
# tick labels are set in alphabetical order
axs1.set_xticklabels(["cat", 'dog', 'mouse', 'bat', 'tree'])

plt.show()
```

```python
# this will create an empty figure and axes object
fig, ax = plt.subplots()

# setting figure size
fig.set_size_inches(3,3)

# customizing plot using axes object
ax.set_xlabel('X label goes here')
ax.set_ylabel('Y label goes here')
ax.set_xticks([0,1,2,3,4,5])
ax.set_xticklabels(['a', 'b', 'c', 'd', 'e', 'f'])

plt.show()
```

```python
# the previous fig and ax objects can no longer be used due to the finalization of the previous plt.show() call
# thus we need to create new fig and ax objects
fig, ax = plt.subplots()

fig.set_size_inches(6,6)
fig.set_facecolor("#004675")

ax.set_facecolor("#004675")
ax.set_xlabel(xlabel='x', color="#ffffff")
ax.set_ylabel(ylabel='y', color="#ffffff")
ax.tick_params(axis='x', colors="#ffffff")
ax.tick_params(axis='y', colors="#ffffff")
ax.spines[:].set_color("#ffffff")

x = np.linspace(-2*np.pi, 2*np.pi, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

line1, = ax.plot(x, y1, color='#c30000', linestyle='-')
line2, = ax.plot(x, y2, color='#ffd22e', linestyle='--')
ax.legend([line1, line2], ['y = sin(x)', 'y = cos(x)'], loc='upper right')
ax.grid(visible=True, which='major', color="#ffffff", linewidth=0.25)
ax.set_xlim(left=0, right=6)
ax.set_title("Multiple Plots", color="#ffffff")

plt.show()
```

### 2 Line Plots
---

```python
# initial plot setup
fig, ax = plt.subplots()

fig.set_size_inches(6,6)
fig.set_facecolor("#004675")

ax.set_facecolor("#004675")
ax.set_xlabel(xlabel='x', color="#ffffff")
ax.set_ylabel(ylabel='y', color="#ffffff")
ax.tick_params(axis='x', colors="#ffffff")
ax.tick_params(axis='y', colors="#ffffff")
ax.spines[:].set_color("#ffffff")
ax.grid(visible=True, linewidth=0.15, color="#ffffff")
ax.set_title("Annotated Plot", color="#ffffff")

# generating and plotting data points
x = np.linspace(-1,1,21)
y1 = np.abs(np.random.randn(21))
ax.plot(x, y1, '--o', color='#ffd22e')

# here we add an annotation to the midpoint of our data
ax.text(x[10], y1[10], 'midpoint', color="#ffffff", fontsize=14)

plt.show()
```

```python
# initial plot setup
fig, ax = plt.subplots()

fig.set_size_inches(6,6)
fig.set_facecolor("#004675")

ax.set_facecolor("#004675")
ax.set_xlabel(xlabel='x', color="#ffffff")
ax.set_ylabel(ylabel='y', color="#ffffff")
ax.tick_params(axis='x', colors="#ffffff")
ax.tick_params(axis='y', colors="#ffffff")
ax.spines[:].set_color("#ffffff")
ax.grid(visible=True, linewidth=0.15, color="#ffffff")
ax.set_title("Filled Plot", color="#ffffff")

# this code plots a line and only a line
ax.plot(x, y1, linestyle='-', marker='o', color='#ffd22e')
# this code fills a region bounded above by the line defined by points x and y1
ax.fill_between(x, y1, color='#ffd22e', alpha=0.1)

plt.show()
```

```python
# initial plot setup
fig, ax = plt.subplots()

fig.set_size_inches(10,6)
fig.set_facecolor("#004675")

ax.set_facecolor("#004675")
ax.set_xlabel(xlabel='x', color="#ffffff")
ax.set_ylabel(ylabel='y', color="#ffffff")
ax.tick_params(axis='x', colors="#ffffff")
ax.tick_params(axis='y', colors="#ffffff")
ax.spines[:].set_color("#ffffff")
ax.grid(visible=True, linewidth=0.15, color="#ffffff")
ax.set_title("Multi-Line Filled Plot", color="#ffffff")

# the line y=0
y0 = np.zeros_like(x)

# this function will vertically compress and reflect y1 about the line y=0
y2 = - y1 / 4

# plotting the lines
line1, = ax.plot(x, y1, color="#c00000", linestyle='-', marker='o')
line2, = ax.plot(x, y2, color="#64c100", linestyle='-', marker='o')
line3, = ax.plot(x, y0, color="#000000", linestyle='-', marker='', linewidth=0.5)

# filling region bounded above by y1(x) and below by y0(x)
ax.fill_between(x,y1,y0,color="#c00000", alpha=0.1)
# filling region bounded above by y0(x) and below by y2(x)
ax.fill_between(x,y0,y2,color="#64c100", alpha=0.1)

ax.legend([line1, line2, line3], [r'$y1 = |\{Y_{i}\}^{21}_{i=1} \sim N(0,1)|$', r'$y2 = \frac{-y1}{4}$', r'$y0 = 0$'], fontsize=14)

plt.show()
```

### 3 Bar Plots
---

```python
# initial plot setup
fig, ax = plt.subplots()

fig.set_size_inches(6,4)
fig.set_facecolor("#004675")

ax.set_facecolor("#004675")
ax.set_xlabel(xlabel='student', color="#ffffff")
ax.set_ylabel(ylabel='grade', color="#ffffff")
ax.tick_params(axis='x', colors="#ffffff")
ax.tick_params(axis='y', colors="#ffffff")
ax.spines[:].set_color("#ffffff")
ax.grid(visible=True, linewidth=0.15, color="#ffffff")
ax.set_title("Bar Plot", color="#ffffff")

# plot data
# xtick labels
names = ['Jack', 'Jill', 'Joe']
# xticks locations
x = np.array([0, 1, 2])
# y-axis data points
grades = [70, 75, 80]

ax.bar(x, grades, color="#ffda45ff", width=0.25)
# the tick locations and labels can be set at once
ax.set_xticks(x, names)

plt.show()
```

```python
# initial plot setup
fig, ax = plt.subplots()

fig.set_size_inches(6,6)
fig.set_facecolor("#004675")

ax.set_facecolor("#004675")
ax.set_xlabel(xlabel='student', color="#ffffff")
ax.set_ylabel(ylabel='grades', color="#ffffff")
ax.tick_params(axis='x', colors="#ffffff")
ax.tick_params(axis='y', colors="#ffffff")
ax.spines[:].set_color("#ffffff")
ax.grid(visible=True, linewidth=0.15, color="#ffffff", zorder=0)
ax.set_title("Multi-Bar Plot", color="#ffffff")

# plot data
# xtick labels
names = ['Jack', 'Jill', 'Joe']
# xticks locations
x = np.array([0, 1, 2])
# y-axis data points
grades_fall = grades
grades_winter = [95, 95, 90]

# plotting bars
    # axis view will automatically be extended to accomodata for the set x-position of the bars
bar1 = ax.bar(x-0.12, grades_fall, color="#229100ff", width=0.24, zorder=10)
bar2 = ax.bar(x+0.12, grades_winter, color="#ff3c2aff", width=0.24, zorder=10)

# the tick locations and labels can be set at once
ax.set_xticks(x, names)

# legend
    # customizing the legend requires returning the legend object and modifying its properties
legend = ax.legend([bar1, bar2], ['fall grades', 'winter grades'], loc='right')
legend.get_frame().set_facecolor(color='#525252')
legend.get_frame().set_edgecolor(color='#525252')
legend.set_zorder(20)
legend.get_frame().set_alpha(0.8)
for t in legend.get_texts():
    t.set_color('#ffffff')

plt.show()
```

```python
# initial plot setup
fig, ax = plt.subplots()

fig.set_size_inches(6,6)
fig.set_facecolor("#004675")

ax.set_facecolor("#004675")
ax.set_xlabel(xlabel='student', color="#ffffff")
ax.set_ylabel(ylabel='grades', color="#ffffff")
ax.tick_params(axis='x', colors="#ffffff")
ax.tick_params(axis='y', colors="#ffffff")
ax.spines[:].set_color("#ffffff")
ax.grid(visible=True, linewidth=0.15, color="#ffffff")
ax.set_title("Stacked Bar Plot", color="#ffffff")

# plot data
# xtick labels
names = ['Jack', 'Jill', 'Joe']
# xticks locations
x = np.array([0, 1, 2])
# can set tick values and labels at once
ax.set_xticks(x, names)
# data points
grades = np.array([
    [30, 30, 10],
    [50, 30, 15],
    [45, 23, 5],
])

# bottom refers to the bottom of the bar
    # the bottom of each bar should be the top of the previous bar

# the bottom of these bars is just y=0
ax.bar(x, grades[:, 0], label='final', color="#059d00", zorder=2)
# the bottom of these bars is the top of grades[:,0]
ax.bar(x, grades[:, 1], bottom=grades[:, 0], label='midterm', color="#b02f00", zorder=2)
# here the bottom needs to be calculated, we sum the columns of the array grades[:,:2] using .sum(axis=1)
ax.bar(x, grades[:, 2], bottom=grades[:, :2].sum(axis=1), label='quiz', color="#9600a1", zorder=2)
legend = ax.legend()
legend.get_frame().set_facecolor("none")
legend.get_frame().set_edgecolor("#ffffff")
for t in legend.get_texts():
    t.set_color("#ffffff")
plt.show()
```
